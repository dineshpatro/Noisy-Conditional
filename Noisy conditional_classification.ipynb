```python
import numpy as np
from __future__ import print_function
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
#Input feature
x = np.linspace(0,1,100)

#noise
np.random.seed(555)
noise = np.random.uniform(-0.2,0.2,100)

#Target variable
y = ((x + noise) > 0.5).astype(int)
```

```python
x.shape
```

```python
X = x.reshape(100,1)
```

```python
print (X[:5])
```

```python
plt.scatter(X,y)
plt.show()
```

<ol>
<li>If x>0.5, we want our model to predict higher probability for y=1.</li>
<li>If xâ‰¤0.5, we want our model to predict higher probability for y=0.</li>
<li>If x is closer to 0.5, we want less confidence in the prediction.</li>
<li>If x is farther from 0.5, we want more confidence in the prediction.</li>
</ol>

```python
model = LinearRegression()
model.fit(X,y)


plt.scatter(X,y)
plt.plot(X, model.predict(X), 'k--')
plt.show()
```

```python
model = LogisticRegression()
model.fit(X,y)
pred = model.predict_proba(X[:10])

#Result is a list of lists. 
print ("positive class for first observation is ", pred[0][1])
```

```python
def fit_and_plot_classifier(clf):
    model = clf
    model.fit(X,y)
    pred = model.predict_proba(X)

    pred = [p[1] for p in pred]

    plt.scatter(X,y)
    plt.plot(X, pred, 'k--')
    plt.show()
    return model, pred
```

```python
clf, pred = fit_and_plot_classifier(LogisticRegression(penalty='l1', C = 4))
```

```python
clf, pred = fit_and_plot_classifier(RandomForestClassifier(n_estimators=50))
```
